```python
import time
import pandas as pd
from google import genai
from google.colab import userdata
```

```python
api_key = userdata.get('GEMINI_API')
client = genai.Client(api_key=api_key)
model="gemini-2.5-flash"
```

#**Extract**

```python
df = pd.read_csv('clientes.csv')
print(df)
```

# **Transform**

```python
def gerar_mensagem(cliente):
    prompt = (
        f"Aja como um redator publicitário bancário."
        f"Escreva APENAS uma frase curta (máximo 90 caracteres) para {cliente['Nome']} "
        f"sobre a importância de investir no banco {cliente['Banco']}. "
        f"não use quebras de linha e não dê explicações. Vá direto ao ponto. "
        )

    try:
        time.sleep(5)
        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=prompt
        )

        return response.text.strip().replace('"', '')

    except Exception as e:
        return "Invista no seu futuro!"


# Execução
print("Gerando mensagens...")
df['MensagemPersonalizada'] = df.apply(gerar_mensagem, axis=1)

# Mostrar resultado
print(df[['Nome', 'MensagemPersonalizada']])
```

# **LOAD**

```python
# Salva o resultado em um arquivo CSV
df.to_csv('marketing.csv', index=False, encoding='utf-8-sig')


from google.colab import files
files.download('marketing.csv')
```
